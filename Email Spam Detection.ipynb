# Spam Email Setection System

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('mail_data.csv')
```

```python
df
```

```python
data = df.where(pd.notnull(df), '')
```

```python
data.head(10)
```

```python
data.info()
```

```python
data.shape
```

```python
data.loc[data['Category'] == 'spam', 'Category', ] = 0
data.loc[data['Category'] == 'ham', 'Category', ] = 1
```

```python
x = data['Message']
y = data['Category']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
print(y.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
feature_extraction = TfidfVectorizer(min_df=1, stop_words= 'english', lowercase=True)

x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.transform(x_test)

y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(x_train)
```

```python
print(x_train_features)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train_features, y_train)
```

```python
prediction_on_training_data = model.predict(x_train_features)
accuracy_on_training_data = accuracy_score(y_train, prediction_on_training_data)
```

```python
print('Acc on training data: ', accuracy_on_training_data)
```

```python
prediction_on_test_data = model.predict(x_test_features)
accuracy_on_test_data = accuracy_score(y_test, prediction_on_test_data)
```

```python
print('acc on test data : ', accuracy_on_test_data)
```

```python
input_your_mail = ["""Had your mobile 11 months or more? U R entitled to Update to the latest colour mobiles with camera for Free! Call The Mobile Update Co FREE on 08002986030"""]
input_data_features = feature_extraction.transform(input_your_mail)
predication = model.predict(input_data_features)
```

# Spam Email Detection Using Naive Bayes 

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB 
from sklearn.metrics import accuracy_score
```

```python
# import data
spam_df = pd.read_csv("mail_data.csv")
print(spam_df.groupby('Category').describe())
```

```python
# create spam column to represet spam and ham emails with 1s and 0s
spam_df['spam'] = spam_df['Category'].apply(lambda x: 1 if x == 'spam' else 0)
```

```python
# Create train and test split
x_train, x_test, y_train, y_test = train_test_split(spam_df.Message, spam_df.spam, test_size=0.25)
```

```python
# find word count and store data as a matrix
cv = CountVectorizer()
x_train_count = cv.fit_transform(x_train.values)
```

```python
# train model
model = MultinomialNB()
model.fit(x_train_count, y_train)
```

```python
# test the model 
x_test_count = cv.transform(x_test.values)
prediction_on_training_data = model.predict(x_test_count)
accuracy_on_training_data = accuracy_score(y_test, prediction_on_training_data)
print(model.score(x_test_count, y_test))
print('acc on test data : ', accuracy_on_test_data)
```

```python
# test the model with real-world email
input_your_mail = ["Hey there, can you  call me"]
input_data_count = cv.transform(input_your_mail)
predication = model.predict(input_data_count)
print(predication)
```

# Spam Email Detection Usin SVM

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import SVC  
from sklearn.metrics import accuracy_score
```

```python
# load data
df_spam = pd.read_csv('mail_data.csv')
print(df_spam)
```

```python
# create spam column to represet spam and ham emails with 1s and 0s
spam_df['spam'] = spam_df['Category'].apply(lambda x: 1 if x == 'spam' else 0)
```

```python
# Create train and test split
x_train, x_test, y_train, y_test = train_test_split(spam_df.Message, spam_df.spam, test_size=0.25)
```

```python
# find word count and store data as a matrix
cv = CountVectorizer()
x_train_count = cv.fit_transform(x_train.values)
```

```python
# test the model 
x_test_count = cv.transform(x_test.values)
prediction_on_training_data = model.predict(x_test_count)
accuracy_on_training_data = accuracy_score(y_test, prediction_on_training_data)
print(model.score(x_test_count, y_test))
print('acc on test data : ', accuracy_on_test_data)
```

```python
# test the model with real-world email
input_your_mail = ["Hey there, can you  call me"]
input_data_count = cv.transform(input_your_mail)
predication = model.predict(input_data_count)
print(predication)
```

# Spam Mail Detection Using Ensemble

```python
import string

import pandas as pd
import numpy as np

import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
nltk.download('stopwords')
```

```python
df = pd.read_csv('spam_ham_dataset.csv')
```

```python
df['text'] = df['text'].apply(lambda x: x.replace('\r\n', ''))
```

```python
df.info()
```

```python
stemmer = PorterStemmer()
corpus = []

stopwords_set = set(stopwords.words('english'))

for i in range(len(df)):
    text = df['text'].iloc[i].lower()
    text = text.translate(str.maketrans('', '', string.punctuation)).split()
    text = [stemmer.stem(word) for word in text if word not in stopwords_set ]
    text = ' '.join(text)
    corpus.append(text)
```

```python
vectorzer = CountVectorizer()

x = vectorzer.fit_transform(corpus).toarray()
y = df.label_num

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
clf = RandomForestClassifier(n_jobs=-1)

clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```
